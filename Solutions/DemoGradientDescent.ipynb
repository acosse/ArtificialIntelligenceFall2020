### Demo gradient descent +  regression on non linear data

#### Demo 1. Gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt


# generating the data 

numPoints = 20

x = np.linspace(0,1,numPoints)

beta0 = 1
beta1 = 1

y = beta0 + beta1*x

noise = np.random.normal(0,0.1,numPoints)

ynoisy = y +  noise 


plt.plot(x, y)
plt.scatter(x, ynoisy, c='r')
plt.show()

```

```python
# gradient descent

maxIter = 200

# loss  = sum (y_i - (beta0 + beta1*x_i))^2

# derivative_beta0 = -2 sum (y_i - (beta0 + beta1 x_i)) * 1
# derivative_beta1 = -2 sum (y_i - (beta0 + beta1 x_i)) * x_i

numIter = 0

lr = .01

learned_beta0 = np.random.normal(0,1,1)
learned_beta1 = np.random.normal(0,1,1)

while numIter < maxIter:
    
    
    prediction = learned_beta0 + learned_beta1*x
    
    learned_beta0 += 2*lr*np.sum(ynoisy.reshape(-1,1) - prediction.reshape(-1,1))
    learned_beta1 += 2*lr*np.sum(np.multiply(ynoisy.reshape(-1,1) \
                                             - prediction.reshape(-1,1),x.reshape(-1,1)))
    
    
    numIter += 1

    
xtest = np.linspace(0,1,100)

prediction = learned_beta0 + learned_beta1*xtest
print(learned_beta1)
print(learned_beta0)


plt.plot(xtest, prediction, c='g')
plt.plot(xtest, beta0 + beta1*xtest)
plt.scatter(x, ynoisy, c='r')
plt.show()

```
